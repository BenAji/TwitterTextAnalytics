### Sentiment Analysis
This classifies the senato's tweet to be one being either Neutral,Positive or Negative speech

```python
from textblob import TextBlob
import re

def clean_tweet(tweet):
    '''
    Utility function to clean the text in a tweet by removing 
    links and special characters using regex.
    '''
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())

def analize_sentiment(tweet):
    '''
    Utility function to classify the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

#### SA(sentiment Analysis) Table

Below shows the table where SA is appended 

```python
# We create a column with the result of the analysis:
datae['SA'] = np.array([ analize_sentiment(tweet) for tweet in datae['Tweets'] ])

# We display the updated dataeframe with the new column:
display(datae.head(10))
```

```python
# We construct lists with classified tweets:

pos_tweets = [ tweet for index, tweet in enumerate(datae['Tweets']) if datae['SA'][index] > 0]
neu_tweets = [ tweet for index, tweet in enumerate(datae['Tweets']) if datae['SA'][index] == 0]
neg_tweets = [ tweet for index, tweet in enumerate(datae['Tweets']) if datae['SA'][index] < 0]
```

#### SA  Classification Of Tweets

Below we see that 74% of Dino texts are classified neutral and 9% is classified as Negative speech
